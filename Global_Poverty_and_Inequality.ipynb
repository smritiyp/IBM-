# Description of the Dataset


The dataset from the Cross-National Data Center in Luxembourg offers detailed insights into global poverty and inequality. It includes metrics such as the Gini Coefficient and Atkinson Coefficients to measure income inequality, and various percentile ratios to compare income distributions. It also provides relative poverty rates for the total population, children, and the elderly at different thresholds. Additionally, the dataset covers the distribution of children by income groups, poverty rates among children in different family structures, and the percentage of children living in single-mother families. Median and mean equivalized incomes are also included, making it a comprehensive resource for analyzing income disparities and poverty levels across diverse demographics.

**Column Names with their meanings**
*   gini - Gini Coefficient
*   atk5 - Atkinson Coefficient (epsilon=0.5)

*   atk1 - Atkinson Coefficient (epsilon=1)
*   d9010 - Percentile Ratio (90/10)

*   d9050 - Percentile Ratio (90/50)
*   d8020 - Percentile Ratio (80/20)


*   poorAll4 - Relative Poverty Rates - Total Population (40%)
*   poorAll5 - Relative Poverty Rates - Total Population (50%)

*   poorAll6 - Relative Poverty Rates - Total Population (60%)
*   poorK4 - Relative Poverty Rates - Children (40%)

*   poorK5 - Relative Poverty Rates - Children (50%)
*   poorK6 - Relative Poverty Rates - Children (60%)

*   poorE4 - Relative Poverty Rates - Elderly (40%)
*   poorE5 - Relative Poverty Rates - Elderly (50%)

*   poorE6 - Relative Poverty Rates - Elderly (60%)
*   d5075 - Distribution of Children by Income Group (50-75%)

*   d75150 - Distribution of Children by Income Group (75-150%)

*   d150 - Distribution of Children by Income Group (above 150%)

*   poortp - Children Poverty Rates - Two-Parent Families (50%)
*   poorsm - Children Poverty Rates - Single-Mother Families (50%)


*   pkidsm - % Children Living in Single-Mother Families


*   eymed - Median Equivalized Income

*   average - Mean Equivalized Income

It is the ratio of correctly predicted positive (TP) observations to the total predicted positive (TP+FP) observations.

Precision=TP/(TP+FP)

Where TP=True Positive

# ROC Curve & ROC AUC

ROC curve is one the important evaluating metrics that should be used to check the performance of an classification model.It is also called relative operating characteristic curve, because it is a comparison of two main characteristics (TPR and FPR). It is plotted between sensitivity(aka recall aka True Positive Rate) and False Positive Rate (FPR = 1-specificity).

ROC (Receiver Operating Characteristic) Curve tells us about how good the model can distinguish between two things (e.g If a patient has a disease or no).

Area Under Curve (AUC) helps us to choose the best model amongst the models for which we have plotted the ROC curves

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

# Create a synthetic dataset with fewer classes to avoid mismatch
n_classes = 3
X, y = make_classification(n_samples=1000, n_features=20, n_informative=3, n_redundant=0, n_classes=n_classes, n_clusters_per_class=1, random_state=42)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a logistic regression model
lr_model = LogisticRegression(multi_class='ovr', solver='lbfgs', max_iter=1000)
lr_model.fit(X_train, y_train)

# Predict probabilities for each class
lr_pred_proba = lr_model.predict_proba(X_test)

# Calculate the ROC AUC score for the multi-class classification
auc = roc_auc_score(y_test, lr_pred_proba, multi_class='ovr')

print("ROC AUC SCORE of Logistic Regression is", auc)
```
